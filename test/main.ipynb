```python
##data pre-processing
```

```python
num_cols = [
    "Original Price", "Discount Percentage", "Rating",
    "Number OF Ratings", "Battery Life (Days)",
    "Display Size", "Weight"
]

df[num_cols].dtypes
```

```python
df[["Display Size","Weight"]].dtypes
```

```python
df["Display Size"]=(df["Display Size"].astype("str").str.replace("inch","",regex=False).str.strip())
df["Display Size"]=pd.to_numeric(df["Display Size"],errors="coerce")
```

```python
df[["Display Size"]].dtypes 
```

```python
df["Weight"]=(df["Weight"].astype("str").str.replace("g","",regex=False).str.strip())
df["Weight"]=pd.to_numeric(df["Weight"],errors="coerce")
```

```python
df[["Weight"]].dtypes 
```

```python
df[num_cols].dtypes
```

```python
df.dtypes
```

```python
df.drop(columns=["Model Name"],inplace=True)
```

```python
num_cols=df.select_dtypes(include='number').columns
```

```python
df[num_cols]=df[num_cols].fillna(df[num_cols].mean()[0])
df.head()
```

```python
cato_columns=df.select_dtypes(include='object').columns
```

```python
for col in cato_columns:
    df[col]=df[col].fillna(df[col].mode()[0])
```

```python
binary_col=["Touchscreen","Bluetooth"]
binary_col
```

```python
for col in binary_col:
    df[col]=df[col].map({"Yes":1,"No":0})
```

```python
df.head()
```

```python
cat_cols = df.select_dtypes(include='object').columns
cat_cols
```

```python
brand_unique_values=df["Brand"].nunique()
shape_unique_values=df["Dial Shape"].nunique()
color_unique_values=df["Strap Color"].nunique()
strap_unique_values=df["Strap Material"].nunique()
print("brand_unique_values ",brand_unique_values)
print("shape_unique_values ",shape_unique_values)
print("color_unique_values ",color_unique_values)
print("strap_unique_values ",strap_unique_values)
```

```python
df = pd.get_dummies(df, columns=cat_cols, drop_first=True)
```
